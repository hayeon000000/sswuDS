### 1.

- 개방 주소 방법 : 충돌 발생 시 비어 있는 다른 버킷을 찾아 데이터를 저장하는 방식
    - 장점: 별도의 메모리 공간을 사용하지 않아 메모리 효율이 높다.
    - 단점: 적재율이 높아질수록 성능 저하. 데이터 삭제 시 'DELETE' 표시해야 함

- 체이닝 : 각 버킷에 연결 리스트 두어 충돌된 데이터를 리스트에 추가하는 방식
    - 장점: 구현이 비교적 간단, 적재율이 1을 넘어가도 성능 저하가 완만. 삽입 삭제가 매우 직관적
    - 단점 : 연결 리스트를 위한 추가적인 메모리 필요


### 2. 

1. 

- 결점1 - 공간 낭비 : 해시 테이블 크기는 1,204인데 해시 함수가 1,024를 사용하면, 인덱스 1024부터 1203까지의 공간은 사용되지 않아 메모리가 낭비됨

- 결점2 - 비트 미활용 : 1024로 나눈 나머지는 키값의 하위 10비트만을 결정 인자로 사용. 만약 키값들의 하위 비트가 유사한 패턴을 가진다면 충돌이 빈번히 발생할 수 있음. 일반적으로는 소수를 테이블 크기로 잡는 것이 충돌 줄이는 데 유리함.

2. 이유 : 해시 함수는 항상 동일한 입력에 대해 동일한 해시값을 반환해야 함.

4. [1: 40, 4: 30, 7: 20, 8: 33, 9: 46, 10: 10. 11: 50, 12: 60]

5. [1: 40, 4: 30, 7: 20, 8: 33, 9: 60, 10: 10, 11: 46, 12: 50]

6. [0: 46, 1: 40, 4: 30, 6: 60, 7: 20, 8: 33, 10: 10, 11: 50]

7. 맨 앞에 삽입하면 항상 O(1)의 시간이 걸리지만, 맨 뒤에 삽입하면 리스트 전체를 순회해야 하므로 O(n)의 시간 소요

8. 
- 삽입: 단순히 맨 앞에 넣는 것보다 효율성이 떨어짐. 삽입 위치를 찾지 위해 O(n)의 순회 시간이 필요하기 때문.
- 삭제: 리스트를 순회하며 노드를 찾아야 하므로 O(n)이 소요되나, 찾는 값보다 큰 값을 만나면 즉시 중단할 수 있어 평균적으로 약간 빨라짐.
- 성공적인 검색: O(n)으로 큰 변화가 없음.
- 실패하는 검색: 크게 향상됨. 전체 리스트를 끝까지 볼 필요 없이 검색 키보다 큰 값의 노드에 도달하면 즉시 실패로 처리할 수 있음.
