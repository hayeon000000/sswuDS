### 03
문제 가장 왼쪽의 리프 노드 = 1

1번. 5, 3, 1, 8, 7, 6, 9

2번. 1, 3, 5, 6, 7, 8, 9

3번. 1, 3, 6, 7, 9, 8, 5

4번. 
    베이스 케이스 : 높이 h = 1
    이진 검색 트리의 유일한 노드(루트 노드)

    귀납적 가정 : k<h인 모든 k에 대해 성립한다고 가정.

    귀납적 전개 : 
        이진 검색 트리에서, 귀납적 가정에 의해 중위 순회는 루트 노드 기준 왼쪽 서브 트리와 오른쪽 서브 트리를 각각 정렬된 순서로 방문한다.
        왼쪽 서브 트리의 모든 키는 루트 노드의 키보다 작고, 오른쪽 서브 트리의 모든 키는 루트 노드의 키보다 크므로
        중위 순회 '왼쪽 서브 트리 -> 루트 -> 오른쪽 서브 트리'는 정렬된 순서이다.


### 04

루트 55 > 30 이므로 왼쪽으로 이동 -> 15

15 < 30 이므로 오른쪽으로 이동 -> 28

28 < 30 이므로 오른쪽으로 이동 -> 30

30 = 30 이므로 노드 찾음!
